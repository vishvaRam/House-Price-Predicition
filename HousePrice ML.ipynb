# Importing

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt\
# for plt bg
plt.rcParams['axes.facecolor'] = 'white'
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

# Reading Dataset

```python
dataSet = pd.read_csv("housePricing.csv")
dataSet.info()
```

```python
dataSet.isnull().sum()
```

```python
df =dataSet.dropna( axis=1)
```

## Dealing with null values

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop(["PID","Order","Misc Val"],axis=1,inplace=True)
```

```python
df.info()
```

# Visualizing

```python
sns.set(style='whitegrid', palette="deep", font_scale=1.2, rc={"figure.figsize": [10, 6]})
sns.histplot(data= df.SalePrice,bins=20,kde=True)
```

### Numarical 

```python
 numarical = df.select_dtypes(np.number);
```

```python
numarical.info()
```

### Visualizing Numarical Values

```python
# Padding


fig, ax = plt.subplots(5,5,figsize=(30,20))
ax = ax.flatten()
index =0
for col,value in numarical.items():
    plt.tight_layout(pad=10,h_pad=1,w_pad=0.7)
    sns.distplot(value,bins=20,ax=ax[index]);
    index += 1
```
